```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.sankey import Sankey

```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, xticks=[], yticks=[], title="Flow Diagram of a Widget")
sankey = Sankey(ax=ax, scale=0.01, offset=0.2, head_angle=180, format='%.0f', unit='%')

sankey.add(flows=[25, 20, 60, 10, 20, 5, 15, 10, 40],
           labels=['', '', '', 'First', 'Second', 'Third', 'Fourth', 'Fifth', 'Hurray!'],
           orientations=[1, 1, 1, 1, 1, 1, 1, 1, 1],
           pathlengths=[0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25],
           patchlabel="Widget\nA")  
# Arguments to matplotlib.patches.PathPatch()

sankey.add(flows=[25, 20, 60, 10, 20, 5, 15, 10, 40],
           labels=['', '', '', 'First', 'Second', 'Third', 'Fourth', 'Fifth', 'Hurray!'],
           orientations=[0, 0, 0, 0, 0, 0, 0, 0, 0],
           pathlengths=[0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25],
           patchlabel="Widget\nA",
           connect=(0,0)
          )  

diagrams = sankey.finish()
diagrams[0].texts[-1].set_color('r')
diagrams[0].text.set_fontweight('bold')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, xticks=[], yticks=[], title="Two Systems")
flows = [0.25, 0.15, 0.60, 0.10, 0.05, 0.25, 0.15, 0.10, 0.35]
sankey = Sankey(ax=ax, unit=None)
sankey.add(flows=flows, label='one',
           orientations=[0, 0, 0, 0, 0, 0, 0, 0, 0])

diagrams = sankey.finish()
diagrams[-1].patch.set_hatch('/')
plt.legend(loc='best')
plt.show()
```
